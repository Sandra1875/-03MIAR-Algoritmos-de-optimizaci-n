**(*) Cuantas posibilidades hay sin tener en cuenta las restricciones?**<br>Respuesta:

Si no hay restricciones sobre cuantas tomas se pueden realizar diariamente tenemos que del conjunto de 30 tomas tomamos **k** elementos donde k varia de 1 a 30. Este caso se trata de una combinación sin repeticion, y el **numero de posibilidades** es:  <br> 
$\sum _{k=1}^n\left[\frac{n!}{k!\cdot \left(n-k\right)!}\right]$
<br>
$\sum _{k=1}^{30}\left[\frac{30!}{k!\cdot \left(30-k\right)!}\right]$ = $\left[\frac{30!}{1!\cdot \left(30-1\right)!}\:+\frac{30!}{2!\cdot \:\left(30-2\right)!}+...+\frac{30!}{30!\cdot \:\left(30-30\right)!}\:\right]$

**¿Cuantas posibilidades hay teniendo en cuenta todas las restricciones.**<br>Respuesta:
<br>La **restricción** es que se puede rodar máximo 6 tomas por día. Como queremos minimizar los costos, vamos a fijar al número de tomas diarias en 6. El problema corresponde a una combinación sin repetición, donde de 30 tomas se elige conjuntos de 6 tomas por día, teniendo en cuenta que no se puede elegir más de una vez un elemento y que, el orden de elección de tomas en un día no importa. El **número de posibilidades** es:
<br>$\frac{n!}{k!\cdot \left(n-k\right)!}$  = $\frac{30!}{6!\cdot \left(30-6\right)!}$ = 593.775

```
import numpy as np
```

```
def Calcular_Costo(Matriz):
  # calculemos el costo 
  No_actores_diario=0
  No_actores_total=0
  for i in range(5):
    Actores = Tomas_df.iloc[Matriz[i]]
    for k in range(9):
      Cantidad_actor = Actores.iloc[:,k:k+1].sum()
      if(Cantidad_actor.iloc[0]>0):
        No_actores_diario+=1
    No_actores_total+=No_actores_diario
    No_actores_diario=0
  return No_actores_total
```

```
def Genera_tomas_diarias(tomas):
  toma_inicial =0
  Matriz = np.full((5,6),0)
  # para los cinco (5) dias
  for dia in range(5):
    for i in range(6):
      Matriz[dia][i]=tomas.index[np.random.randint(0,len(tomas))]
      tomas.drop(Matriz[dia][i], inplace = True)
  return Matriz
```

```
def Fuerza_Bruta(NoIntentos):
  MejorCosto =10e10
  MejorSolucion =[]
  for i in range(NoIntentos):
    tomas_tmp = Tomas_df.copy(deep=True)
    Distribucion= Genera_tomas_diarias(tomas_tmp)
    Costo = Calcular_Costo(Distribucion)
    # funcion objetivo, encontrar la combinacion con el menor costo.
    if (Costo < MejorCosto):
      MejorCosto= Costo
      MejorSolucion = Distribucion
  display("Mejor Costo : " + str(MejorCosto))
  return MejorSolucion
```

```
from time import time
tiempo_inicio = time()
# Ejecucion del algoritmo po Fuerza Bruta
Solucion= Fuerza_Bruta(1000)
tiempo_transcurrido = time() - tiempo_inicio
print("Tiempo Transcurrido : %0.10f segundos." % tiempo_transcurrido)
display("Distribución de tomas por día : ")
for dia in range(5):
    print("Dia No "+ str(dia+1))
    display(Tomas_df.iloc[Solucion[dia,:]])
```

**(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta.** Argumenta porque crees que mejora el algoritmo por fuerza bruta

Respuesta
<br>
El nuevo algoritmo es mejor que fuerza bruta, ya que no busca de forma aleatoria como llenar las seis tomas en los cinco días y luego calcular el costo para minimizar, sino que llena la matriz de tomas por día, buscando las tomas que comparten los mismos actores que la primera toma seleccionada.
<br>Considerando que para medir cómo de bueno es nuestro algoritmo se debe tomar en cuenta estos dos factores:
1.   Eficiencia en tiempo.
2.   Eficiencia en espacio
<br>Se calculo el tiempo de ejecución del algoritmo de fuerza bruta obteniendo como resultado el valor de Mejor Costo : 32 y Tiempo Transcurrido : 29.6821119785 seconds.
<br>Mientras que para el algoritmo mejor que fuerza bruta que utiliza como eurística el criterio de buscar las tomas con actores similares a la seleccionada inicialmente se obtuvo los siguientes resultados: Costo : 32 y
Tiempo Transcurrido : 0.2151455879 segundos. 
<br>**Los costos son identicos, pero el tiempo de ejecución es 130 veces menor; por tanto, es mejor que el algoritmo de fuerza bruta**  

```
def Mejor_que_fuerza_bruta(tomas_disponible):
  Matriz = np.full((5,6),0)
  # llenamos la matriz por dia, las seis tomas
  toma_inicial =0
  # para los cinco (5) dias
  for dia in range(5):
    toma_inicial =tomas_disponible.index[0]
    a = tomas_disponible.iloc[toma_inicial:toma_inicial+1,:]
    actores_similares = []
    # calculamos las tomas que se parecen más a la toma inicial, en función de los actores involucrados.
    for i in range(1,len(tomas_disponible)):
      b= tomas_disponible.iloc[i:i+1,:]
      d = np.logical_not(np.logical_xor(a, b))
      cantidad = a[d].count().sum()
      actores_similares.append([cantidad, tomas_disponible.index[i]])
    actores_similares.sort(reverse=True)
    Matriz[dia][0]=toma_inicial 
    tomas_disponible.drop(toma_inicial, inplace = True)
    # tomamos las cinco tomas mejores en comparación con la primera seleccionada
    for i in range(5):
      Matriz[dia][i+1]=actores_similares[i][1]
      tomas_disponible.drop(Matriz[dia][i+1], inplace = True)
  return Matriz
```

```
tiempo_inicio = time()
# Ejecucion del algoritmo po Fuerza Bruta
tomas_copy = Tomas_df.copy(deep=True)
Solucion= Mejor_que_fuerza_bruta(tomas_copy)
Costo = Calcular_Costo(Solucion)
display("Costo : " + str(Costo))
tiempo_transcurrido = time() - tiempo_inicio
print("Tiempo Transcurrido : %0.10f segundos." % tiempo_transcurrido)
display("Distribución de tomas por día : ")
for dia in range(5):
    print("Dia No "+ str(dia+1))
    display(Tomas_df.iloc[Solucion[dia,:]])
```
